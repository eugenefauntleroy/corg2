```python
import cv2
import numpy as np
import matplotlib.pyplot as plot
```

```python
img = cv2.imread('cardigan-welsh-corgi.jpeg')
img #show numpy array
```

```python
type(img) #show type
```

```python
img.shape #width, heighth, color channel
```

```python
plot.imshow(img)
```

```python
normal = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plot.imshow(normal) #normalize color order
```

```python
def get_image(image, cmap=None):
    plot.axis('off')
    plot.imshow(image, cmap=cmap)
    plot.show() #gets rid of side values
```

```python
get_image(normal) #executes above function
```

```python
#colorized 
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
get_image(hsv)
```

```python
#grey scale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
get_image(gray, cmap='gray')
```

```python
#blurred

width, height =  normal.shape[:2]
print(width)
print(height)
#check dimension 

if width > 500:
    k = (50,50)
elif width > 200 and width <=500:
    k = (25,25)
else:
    k = (10,10)

#k = (2, 2) adjust here for bluriness 
blur = cv2.blur(normal, k)
get_image(blur)
```

```python
#binary

v, binary = cv2.threshold(gray, 100, 255, cv2.THRESH_BINARY) #<---adjust b&w here
get_image(binary, cmap='gray')

#cute!
```

```python
#invert

invert = cv2.bitwise_not(binary)
get_image(invert, cmap='gray')
```
